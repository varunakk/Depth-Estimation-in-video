```python
import numpy as np
import math
import cv2

def rmsdiff(imageA, imageB):
  err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
  err /= float(imageA.shape[0] * imageA.shape[1])*10
  return math.sqrt(err)
```

```python
er=0
for i in range(70):
  if i<10:
    s="0"+str(i)
  else:
    s=str(i)
  img = cv2.imread('/content/mannequinchallenge/test_data/DAVIS/Annotations/480p/breakdance-flare/000'+s+'.png')
  im2=cv2.imread("/content/mannequinchallenge/test_data/viz_predictions/breakdance-flare/000"+s+".jpg")
  er=er+rmsdiff( cv2.resize(img,(224,224),interpolation = cv2.INTER_AREA), cv2.resize(im2,(224,224),interpolation = cv2.INTER_AREA))
er=er/70
print("RMSE ERROR: ",er)
```
